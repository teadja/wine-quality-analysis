```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Wine Quality Analysis for Production Optimization

This project analyzed wine quality data to identify key factors that influence product consistency and performace


## Objective
The goal of this analysis it to identify the key factors that influence wine quality and provide insights that can support production optimization.

## Data Exploration
Initial exploration is performed to understand the structure of the dataset, check the data types, and summarize key statistics.

```python
df = pd.read_csv( "Wine Quality/winequality-red.csv" , sep=";")
df.head()
```

```python
df["quality"].value_counts().sort_index()
```

## Analysis & Visualizations
Visualizations are used to explore distributions, relationships, and correlations between variables and wine quality.

```python
sns.countplot(x="quality", data=df)
plt.title("Distribution of Wine Quality Ratings")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Wine Feature Correlations")
plt.show()
```

```python
sns.boxplot(x="quality", y="alcohol" , data=df)
plt.title("Alcohol Content vs Wine Quality")
plt.show()
```

```python
df.groupby("quality")["alcohol"].mean()
df.groupby("quality")["volatile acidity"].mean()
```

## Key Insights
- Higher alcohol content is positively associated with higher wine quality ratings, suggesting it may be a key driver of percieved quality
- volatile acidity shows a negative relationship with quality, indicating that lower acidity levels may improve product outcomes
- Most wines fall within mid-range quality scores (5-6), highlighting an opportunity to shift production toward higher-quality segments
-  Variability across chemical properties suggests inconsistency in production, which could be optimized for better standardization

## Business Impact
These findings can support production teams in refining chemical compositions and improving process consistency, ultimately enhancing product quality and customer satisfaction.
